```python
import requests
dir(requests)
import urllib3
from bs4 import BeautifulSoup
url="https://www.goeduhub.com/"
response=requests.get(url)
print(response)
htmlcontent=response.content
soup=BeautifulSoup(htmlcontent,"html.parser")
#print(soup.prettify) #will print the whole content
```

```python
#title of the page

print(soup.findAll("title"))
```

```python
# will print the text files
print(soup.findAll("text"))
```

```python
#if we want to give an anchor tag to the fourth number inthe webpage

one_tag=soup.findAll('a')[4]
print(one_tag)
```

```python
#how tags are connected to each other

anchor=soup.findAll('a')
all_link=set()
for link in anchor:
    if(link.get('href')!='#'):
        linkprint="https://www.goeduhub.com/"+link.get('href')
        all_link.add(link)
        print(linkprint)
```
